```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Dataset\Algerian_forest_fires_dataset.csv")
```

# Data Cleaning 

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
# Remove these entries from the dataset
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int) 
df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float) 
```

```python
df.info()
```

```python
df.to_csv("Dataset\Algerian_forest_fires_cleaned_dataset.csv", index=False)
```

```python
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# not fire --> 0
# fire     --> 1
def convertClassToBoolean(class_name):
    if 'not' in class_name:
        return 0
    else:
        return 1
df_copy['Classes'] = df_copy['Classes'].apply(convertClassToBoolean)
```

```python
df_copy['Classes'].value_counts()
```
